```python
import pandas as pd

# Load email master list
email_df = pd.read_csv('/mnt/home/elli/workspaces/roomdraw/database/notebooks/Number Master list, Room Draw 2026 - For Digidraw - Email.csv')

# Normalize: convert @hmc.edu -> @g.hmc.edu, lowercase
email_df['Email address'] = email_df['Email address'].str.replace('@hmc.edu', '@g.hmc.edu', regex=False).str.lower()

# Keep only ID and email
id_to_email = email_df[['ID', 'Email address']].rename(columns={'Email address': 'Email address'})

print(f'Loaded {len(id_to_email)} email entries')
id_to_email.head()
```

```python
for filename in ['numbers_juniors.csv', 'numbers_seniors.csv', 'numbers_sophomores.csv']:
    df = pd.read_csv(filename).drop(columns=['Email', 'Email address'], errors='ignore')
    merged = df.merge(id_to_email, on='ID', how='left').rename(columns={'Email address': 'Email'})

    unmatched = merged[merged['Email'].isna()][['ID', 'LastName', 'FirstName']]
    print(f'--- {filename}: {len(unmatched)} unmatched ---')
    if len(unmatched) > 0:
        print(unmatched.to_string(index=False))
    print()

    merged.to_csv(filename, index=False)
    print(f'Saved {filename}\n')
```
