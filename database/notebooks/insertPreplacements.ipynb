```python
# read in preplacements.csv into a dataframe

import pandas as pd
import numpy as np

df = pd.read_csv('preplacements.csv')


# Dorm Mapping
# 1 = East
# 2 = North
# 3 = South
# 4 = West
# 5 = Atwood
# 6 = Sontag
# 7 = Case
# 8 = Drinkward
# 9 = Linde
dorm_mapping = {
    'East': 1,
    'North': 2,
    'South': 3,
    'West': 4,
    'Atwood': 5,
    'Sontag': 6,
    'Case': 7,
    'Drinkward': 8,
    'Linde': 9,
    'Garrett House': 10
}

# remove any rows that don't have email, Dorm, or Room
df = df[df['Email'].notna() & df['Dorm'].notna() & df['Room'].notna()]

# remove any uppercase letters
df['Email'] = df['Email'].str.lower()
# if email ends in @hmc.edu, replace with @g.hmc.edu
df['Email'] = df['Email'].str.replace('@hmc.edu', '@g.hmc.edu')

# print the first 5 rows of the dataframe

# strip any whitespace from first name and last name and email and dorm and room
df['First Name'] = df['First Name'].str.strip()
df['Last Name'] = df['Last Name'].str.strip()
df['Email'] = df['Email'].str.strip()
df['Dorm'] = df['Dorm'].str.strip()
df['Room'] = df['Room'].str.strip()

# remove any rows that are not in the dorm mapping
df = df[df['Dorm'].isin(dorm_mapping.keys())]

print(df)
```

```python
from sqlalchemy import create_engine
from sqlalchemy.sql import text

# import env variables
import os
from dotenv import load_dotenv
from pathlib import Path

# import libraries for ssh tunneling
import sshtunnel

dotenv_path = os.path.join(os.getcwd(), '.env')
print(dotenv_path)

load_dotenv(dotenv_path=dotenv_path, verbose=True)

sql_pass = os.environ.get('SQL_PASS')
sql_ip = os.environ.get('SQL_IP')
sql_db_name = os.environ.get('SQL_DB_NAME')
sql_user = os.environ.get('SQL_USER')

tunnel_host = os.environ.get('TUNNEL_HOST')
tunnel_port = os.environ.get('TUNNEL_PORT')
tunnel_user = os.environ.get('TUNNEL_USER')
tunnel_pass = os.environ.get('TUNNEL_PASS')

tunnel = sshtunnel.SSHTunnelForwarder(
    (tunnel_host, int(tunnel_port)),
    ssh_username=tunnel_user,
    ssh_password=tunnel_pass,
    remote_bind_address=(sql_ip, 5432)
)

# After starting the tunnel
tunnel.start()

# Get the local bind port that the tunnel is using
local_port = tunnel.local_bind_port

# Update connection string to use localhost and the tunneled port
CONNSTR = f'postgresql://{sql_user}:{sql_pass}@127.0.0.1:{local_port}/{sql_db_name}'

# Now create your SQLAlchemy engine with this connection string
engine = create_engine(CONNSTR)

# Test the connection
with engine.connect() as connection:
    result = connection.execute(text("SELECT 1"))
    print("Connection successful!")
```

```python
not_found = []
room_mismatches = []

def get_reslife_role(notes):
    valid_values = ['mentor Pull', 'Mentor', 'Proctor Pull', 'Proctor']
    if notes in valid_values:
        return notes
    return 'none'

with engine.connect() as connection:
    for index, row in df.iterrows():
        csv_dorm = row['Dorm']
        csv_room = str(row['Room'])

        # fetch user including their current room assignment
        query = f"SELECT first_name, last_name, year, draw_number, in_dorm, email, preplaced, room_uuid FROM Users WHERE email = '{row['Email']}'"
        result = connection.execute(text(query))
        user = result.fetchone()

        if row['Email'] not in class_mapping:
            print(f"User {row['Email']} not found in roster, skipping")
            not_found.append({'First Name': row['First Name'], 'Last Name': row['Last Name'], 'Email': row['Email'], 'Dorm': csv_dorm, 'Room': csv_room})
            continue

        values = {
            'email': row['Email'],
            'year': class_mapping[row['Email']],
            'draw_number': 0,
            'in_dorm': 0,
            'preplaced': True,
            'reslife_role': get_reslife_role(row['Notes']),
        }

        if user:
            already_preplaced = user[6]  # preplaced column
            current_room_uuid = user[7]  # room_uuid column

            if already_preplaced and current_room_uuid:
                # look up their current room in the rooms table
                room_result = connection.execute(
                    text("SELECT dorm_name, room_id FROM Rooms WHERE room_uuid = :uuid"),
                    {'uuid': str(current_room_uuid)}
                ).fetchone()

                if room_result:
                    db_dorm, db_room = room_result[0], str(room_result[1])
                    if db_dorm != csv_dorm or db_room != csv_room:
                        print(f"MISMATCH for {row['Email']}: DB has {db_dorm} {db_room}, CSV says {csv_dorm} {csv_room} — overriding metadata")
                        room_mismatches.append({
                            'Email': row['Email'],
                            'DB Dorm': db_dorm, 'DB Room': db_room,
                            'CSV Dorm': csv_dorm, 'CSV Room': csv_room,
                        })
                    else:
                        print(f"OK {row['Email']}: room assignment matches ({csv_dorm} {csv_room})")
                else:
                    print(f"WARNING: could not find room {current_room_uuid} for {row['Email']}")
            else:
                print(f"Updating existing user {row['Email']} (was not yet preplaced)")

            connection.execute(text(
                "UPDATE Users SET in_dorm = :in_dorm, preplaced = :preplaced, draw_number = :draw_number, reslife_role = :reslife_role, year = :year WHERE email = :email"
            ), values)
            connection.commit()
        else:
            values['first_name'] = row['First Name']
            values['last_name'] = row['Last Name']
            print(f"Inserting new user {row['Email']}")
            connection.execute(text(
                "INSERT INTO Users (year, first_name, last_name, email, draw_number, preplaced, in_dorm, reslife_role) VALUES (:year, :first_name, :last_name, :email, :draw_number, :preplaced, :in_dorm, :reslife_role)"
            ), values)

    connection.commit()

if not_found:
    pd.DataFrame(not_found).to_csv('not_found.csv', index=False)
    print(f"Wrote {len(not_found)} not-found users to not_found.csv")

if room_mismatches:
    pd.DataFrame(room_mismatches).to_csv('room_mismatches.csv', index=False)
    print(f"Wrote {len(room_mismatches)} room mismatches to room_mismatches.csv")

print("Successfully updated preplacements")
tunnel.stop()
```
